```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal

data = pd.read_csv('../../flight.csv')

#plt.plot(data['0'])
#plt.plot(data['1'])

integrated = []
integral = 0
for i in range(len(data['5'])):
    integral = integral + data['5'][i]
    integrated.append(integral)

fg=1; fs=180

sig = data['5']
sos = signal.butter(2, fg, btype='highpass', fs=fs, analog=False, output='sos')
filtered = signal.sosfilt(sos, sig)


integrated2 = []
integral2 = 0
for i in range(len(data['5'])):
    integral2 = integral2 + filtered[i]
    integrated2.append(integral2)

plt.plot(data['3'])
plt.plot(data['4'])
plt.show()

plt.plot(data['5'])
plt.plot(filtered)
plt.show()

plt.plot(integrated)
plt.plot(integrated2)
plt.show()
```
